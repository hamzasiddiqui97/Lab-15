19. Now you are required to generate a report which shows all the roles in the movie having title "P-Word Pizza". (Write query in SQL. Here you will have to apply JOINs and fetch the information).

21. There is an actor namely "El Viti". You are required to generate a report which shows all the titles of movies and names of respective roles played by the actor "El Viti". (Write query in SQL. Here you will have to apply JOINs on 3 tables and fetch the information).

```python
pd.read_sql_query("SELECT a.name as 'Actor', m.title as 'Movie', r.name as 'Role' FROM actor as a JOIN movie as m JOIN role as r WHERE a.name = 'El Viti'", conn)
```
